```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
current_palette = sns.color_palette()
```

```python
clinical_data = pd.read_csv('clinical.csv')
clinical_data = clinical_data.iloc[:, 1:]
```

Now we want to do visualize correlation between age and other hyperparameters

```python
# last column, Brca1 mutation
inference = clinical_data.iloc[:, -1].values

Y = clinical_data.iloc[:, 0].values

column = ['diameter(mm)', 'followup_time(yr)', 'grade', 'ERp', 'PRp']

fig = plt.figure(figsize = (20,8))

for i in xrange(len(column)):
    
    plt.subplot(1, len(column), i + 1)
    
    X = clinical_data[column[i]].values
    
    for no, k in enumerate(np.unique(inference)):
        plt.scatter(X[inference == k], Y[inference == k], color = current_palette[no], label = str(k))
    
    plt.legend()
    plt.xlabel(column[i])
    plt.ylabel('Age')

fig.tight_layout()
plt.savefig('correlation_brca1.pdf')
plt.show()
```

based on the visualization,

Inference is Brca1 mutation.
